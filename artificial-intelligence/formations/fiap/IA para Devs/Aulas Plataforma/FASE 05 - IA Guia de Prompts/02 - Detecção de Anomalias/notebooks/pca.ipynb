# PCA para Detecção de Anomalias - Análise de Tráfego de Rede

Este notebook demonstra como usar PCA para detectar anomalias em dados de tráfego de rede, identificando picos e quedas bruscas que podem indicar problemas ou ataques.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

## 1. Simulação de Dados de Tráfego

Gerando dados normais de tráfego em 3 portas diferentes e adicionando dados anômalos:

- **Tráfego Normal**: Valores em torno de 50 com variação de 10
- **Anomalia 1**: Pico muito alto (120, 200, 80) - possível ataque de sobrecarga
- **Anomalia 2**: Queda muito baixa (5, 10, 5) - possível falha ou interrupção

```python
np.random.seed(42)

normal_traffic = np.random.normal(loc=50, scale=10, size=(100, 3))
anomalous_traffic = np.array([[120, 200, 80], [5, 10, 5]])
network_traffic = np.vstack([normal_traffic, anomalous_traffic])

```

## 2. Aplicando PCA e Visualizando

Aplicando o PCA para reduzir as 3 dimensões (tráfego nas 3 portas) para 2 componentes principais, facilitando a visualização e identificação de padrões anômalos.

O PCA vai analisar o tráfego nas portas, entender se estão relacionados e criar duas novas variáveis que representam a maior parte da variação dos dados.

```python
pca = PCA(n_components=2)
pca_result = pca.fit_transform(network_traffic)
explained_variance = pca.explained_variance_ratio_
print(f"Variância explicada por cada componente principal: {explained_variance}")

plt.figure(figsize=(10, 6))
plt.scatter(pca_result[:-2, 0], pca_result[:-2, 1], label='Tráfego Normal', alpha=0.7, color='b')
plt.scatter(pca_result[-2:, 0], pca_result[-2:, 1], color='r', label='Anomalias', marker='x', s=100)
plt.title('Redução Dimensional com PCA (Tráfego de Rede)')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.legend()
plt.grid(alpha=0.5)
plt.show()
```
