# Análise de BLEU Score para Avaliação de Textos Gerados

Este notebook demonstra como calcular o BLEU Score para avaliar a qualidade de textos gerados por modelos de IA.

O BLEU Score compara o texto gerado com um ou mais textos de referência, medindo a similaridade entre eles.

## Objetivo

Aprender a avaliar a qualidade de respostas geradas por modelos de IA através da métrica BLEU Score, identificando quão próximas as respostas estão de textos de referência ideais.


## Importação das Bibliotecas

- **nltk**: Biblioteca para processamento de linguagem natural
- **sentence_bleu**: Função para calcular BLEU Score em nível de sentença
- **SmoothingFunction**: Função para suavizar o cálculo em frases curtas


```python
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
import nltk

# Baixar recursos necessários do NLTK (apenas na primeira execução)
try:
    nltk.data.find('tokenizers/punkt')
except LookupError:
    nltk.download('punkt')
```

## Preparar Textos de Referência e Candidato

- **Referência**: Texto ideal ou correto que queremos comparar
- **Candidato**: Texto gerado pelo modelo que queremos avaliar

Os textos precisam ser tokenizados (divididos em palavras) para o cálculo do BLEU Score.


```python
# Texto de referência (ideal)
referencia = ["""
O ácido acético, com fórmula CH₃COOH, é um líquido incolor e corrosivo.
É utilizado na produção de acetato de vinila e como componente principal do vinagre.
""".split()]

# Texto candidato (gerado pelo modelo)
candidato = """
Ácido acético é um líquido incolor com fórmula CH₃COOH.
Usado na fabricação de acetato de vinila e encontrado no vinagre.
""".split()

print("Referência:", ' '.join(referencia[0]))
print("\nCandidato:", ' '.join(candidato))
```

## Calcular BLEU Score

O BLEU Score compara n-gramas (sequências de palavras) entre o texto candidato e a referência.

**SmoothingFunction**: Usada para evitar que frases curtas recebam score zero. O método 4 é uma das opções mais comuns.

**Valor do BLEU Score:**
- **1.0**: Texto idêntico à referência
- **0.0**: Sem similaridade com a referência
- **0.5-0.7**: Boa similaridade
- **< 0.3**: Baixa similaridade


```python
# Usar suavização para evitar score zero em frases curtas
smoothie = SmoothingFunction().method4

# Calcular BLEU Score
bleu_score = sentence_bleu(referencia, candidato, smoothing_function=smoothie)

print(f"BLEU Score: {bleu_score:.4f}")
print(f"BLEU Score (porcentagem): {bleu_score * 100:.2f}%")
```

## Comparar Diferentes Textos Candidatos

Vamos comparar dois textos candidatos diferentes para ver qual está mais próximo da referência.


```python
# Candidato 1: Mais próximo da referência
candidato1 = """
Ácido acético é um líquido incolor com fórmula CH₃COOH.
Usado na fabricação de acetato de vinila e encontrado no vinagre.
""".split()

# Candidato 2: Mais distante da referência
candidato2 = """
Ácido acético é um composto químico.
É usado em várias aplicações industriais.
""".split()

# Calcular BLEU Score para cada candidato
bleu1 = sentence_bleu(referencia, candidato1, smoothing_function=smoothie)
bleu2 = sentence_bleu(referencia, candidato2, smoothing_function=smoothie)

print(f"BLEU Score Candidato 1: {bleu1:.4f} ({bleu1 * 100:.2f}%)")
print(f"BLEU Score Candidato 2: {bleu2:.4f} ({bleu2 * 100:.2f}%)")
print(f"\nCandidato 1 está {((bleu1 / bleu2) - 1) * 100:.1f}% mais próximo da referência")
```

## Interpretação dos Resultados

**BLEU Score alto (próximo de 1.0):**
- Texto gerado está muito próximo da referência
- Indica que o prompt foi bem formulado e o modelo gerou resposta adequada
- Pode indicar que o modelo "entendeu" bem o que foi pedido

**BLEU Score médio (0.3-0.7):**
- Texto gerado tem alguma similaridade com a referência
- Pode indicar que o prompt precisa ser refinado
- Resposta está no caminho certo, mas pode melhorar

**BLEU Score baixo (< 0.3):**
- Texto gerado está distante da referência
- Pode indicar que:
  - O prompt não foi claro o suficiente
  - O modelo não entendeu o contexto
  - É necessário refinar o prompt e testar novamente

**Conclusão**: Use o BLEU Score para comparar diferentes versões de prompts e identificar qual gera respostas mais próximas do que você espera.
